```python
import os
import numpy as np
import xarray as xr

import warnings
warnings.filterwarnings("ignore",category=RuntimeWarning)
```

```python
#Use data URL from opendap link
url = ("https://dapds00.nci.org.au/thredds/dodsC/fx3/gbr1_bgc_GBR1_H2p0_B3p2_Cfur_Dnrt/gbr1_bgc_simple_2023-03-19.nc")
ds = xr.open_dataset(url)
ds
```

```python
#Select variables
variables = [
    'botz',
    'current_speed',
    'current_dir',
    'w',
    'salt',
    'FineSed',
    'tss',
    'NO3',
    'DIP',
    'DIN',
    'PH',
    'TN',
    'TP',
    'dens_sed',
    'wave_dir',
    'weekly_Coral_IN_up',
    'CS_bleach',
    'erdepflux_total'
]
```

```python
var = ds[variables]
var
```

```python
#slice the data here
first_dim_values = ds.k.values
first_dim_values

#Test slice dimension
sliced_ds = var.sel(k_sed=[0])
sliced_ds
```

```python
#Another slice to make xarray smaller
final = sliced_ds.sel(k=slice(0,20))
final
```

```python
#Method 1 -> convert to data frame -> save to csv file
df = final.to_dataframe()
#With sliced data is about 3gb, non sliced data is too large to convert
```

```python
#Export to CSV 
df = df.to_csv("GBR_day1.csv")
#takes about 4 minutes
```
